```python
import glob
import jieba

jieba.load_userdict("./mydict.txt")
with open("./stop_words.txt","r")as f:
    stop_words_list = [w.replace("\n","")for w in f.readlines()]
```

```python
stop_words_list
```

```python
all_str = ""


for file_path in glob.glob("./res_gossiping/*.txt"):
    with open(file_path,"r",encoding="utf-8") as f:
        article_str = f.read().split("---split---")[0]
        all_str += article_str
        
# print(all_str)
        
        
        
#     print(file_path)
```

```python
word_generator = jieba.cut(all_str)
```

```python
word_list = [w for w in word_generator if (len(w) > 1) and (w not in stop_words_list)]
word_list
```

```python
word_list = list(word_generator)
word_list = [w for w in word_list if (len(w) > 1) and (w not in stop_words_list)]

word_list
```

```python
word_count = {}

for word in word_list:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
word_count        
```

```python
word_count_list = [ (k,v) for k , v in word_count.items() ]
word_count_list
```

```python
word_count_list.sort(key=lambda x: x[1] ,reverse=True)
word_count_list
```
